```python
import numpy as np
import pandas as pd
import random 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm  
```

```python
random.seed(88)
```

```python
#Load genes dataframes 
data = pd.read_csv('Models/enrichedOBL/LargeTS/RotatE/TestSet/df_genes_diseases_hg_prediction_models_genes_annotated_disease.tsv',sep='\t')

data_best = pd.read_csv('Models/enrichedOBL/LargeTS/Elkanoto_XGBoost_RotatE/TestSet/df_genes_diseases_hg_prediction_models_genes_annotated_disease.tsv',sep='\t')

data_pos = pd.read_csv('PositiveTrainingSet/df_genes_diseases_hg_IR-relatedGenes_posSet_annotated_disease.tsv',sep='\t')

data_bad = pd.read_csv('Models/enrichedOBL/LargeTS/LocalOutlierFactor_TransE/TestSet/df_genes_diseases_hg_prediction_models_genes_annotated_disease.tsv',sep='\t')

data_topo = pd.read_csv('Models/enrichedOBL/LargeTS/Topology/TestSet/df_genes_diseases_hg_prediction_models_genes_annotated_disease.tsv',sep='\t')
```

```python
data_bad['gene'].unique
```

```python
#only keep genes who have significant p-value
data = data[data['p-value']==0.0]
data_best = data_best[data_best['p-value']==0.0]
data_pos = data_pos[data_pos['p-value']==0.0]
data_bad = data_bad[data_bad['p-value']==0.0]
data_topo = data_topo[data_topo['p-value']==0.0]
```

```python
len(data_pos['gene'].unique())
```

```python
# concatenate dataframes vertically  
df_combined = pd.concat([data, data_best,data_pos,data_bad,data_topo])  
```

```python
# group by 'gene' column and convert 'model' column values into a list  
df_grouped = df_combined.groupby('gene').agg({'model': lambda x: x.tolist()}).reset_index()  
  
# select only 'gene' and 'model' columns  
df_result = df_grouped[['gene', 'model']]  
```
