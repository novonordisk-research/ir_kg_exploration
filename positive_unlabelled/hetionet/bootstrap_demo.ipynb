```python
import pandas as pd
import numpy as np
from anngel.metrics import amri_on_proba_df, mrr_on_proba_df, mr_on_proba_df, bootstrap

proba_df = pd.read_csv('./2_outputs/standard/elkanoto/proba_df.csv')
proba_df = proba_df.query('test')
```

```python
n_rep = 1000
replace = True
size = None

mrrs = bootstrap(
    df=proba_df,
    fun=mrr_on_proba_df,
    n_rep=n_rep,
    replace=replace,
    size=size,
    # kwargs for custom function
    colname_scores = 'RotatE_0',
    colname_y = 'y',
)

mrrs.mean(), mrrs.std()
```

```python
import seaborn as sns
sns.histplot(mrrs)
```
